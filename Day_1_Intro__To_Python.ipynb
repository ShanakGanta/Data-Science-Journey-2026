```python
                                              #Day 1 0f Data Scientist Goal
#1.Identifiers
#Ex:
Village="Patha Hiramandalam"#Village is the Identifier (Variable) and Patha Hiramandalam is the value assigned
Mandal="Hiramandalam" #Mandal is the Identifier (Variable) and Hiramandalam is the value assigned {Identifiers can be variables, Classes, Functions, Modules}
print(Village)
print(Mandal)
#2. Keywords
''' Key words are prebuilt which cannot be changed '''
conditional_statements="IF,ELSE, ELIF"
conditional_control_statements="FOR,WHILE"
other="BREAK,CONTINUE"
print(conditional_statements)
print(conditional_control_statements)
print(other)
```

```python
#3.Dynamic Typing Language
''' Dynamic Typing languages are automatically detects and decides the data types and executes them'''
num1=60
num2="90"
print(type(num1),num1,type(num2),num2)

#4.Variable
'''Variable are often known as containers of python they contains different types data type and their data'''
integer=45
float1=43.23
long1=1243599
double=123.345654
character="a"
print(integer)
print(float1)
print(long1)
print("%0.3f" %double)
print(character)
```

```python
#7.Operators
  #Arthiematic Operators[+,-,*,/,//]
a=5
b=6
result=(a+b,b-a,a*b,b/a,b//a)
print(result)

  #Relational Operators[==,!=,>,<,>=,<=]
a=7
b=10
result=(a==b,a!=b,a>b,a<b,a>=b,a<=b)
print(result)

  #Assignment Operators[=,+=,-=,*=,/=,%=]
a=18
b=23
print(a,b)
a+=b
b+=a
print(a,b)
a-=b
b-=a
print(a,b)
a*=b
b*=a
print(a,b)
a/=b
b/=a
print(a,b)
a%=b
b%=a
print(a,b)

#Logical Operators[AND,OR,NOT]
A=100
B=500
print(a>b and b>a)
print(a>b or b>a)
print(a>b and not (b>a))
```
