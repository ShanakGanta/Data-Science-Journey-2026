```python
subjects = ["Telugu", "English", "Maths", "Physics", "Chemistry"]
total_marks = 0
failed = False

for i in subjects:
    mark = int(input(f"Enter {i} Subject marks: "))
    
    # Logic Correction: Changed 'and' to 'or' for range validation
    if mark < 0 or mark > 100:
        print("Entered marks are invalid. Please enter marks between 0 and 100.")
        failed = True # Invalid marks treat as fail for safety
    else:
        total_marks += mark
        if mark >= 35:
            if mark > 90:
                print('Grade A')
            elif mark > 75:
                print('Grade B')
            elif mark > 60:
                print("Grade C")
            elif mark > 45:
                print("Grade D")
            else:
                print("Grade E")
        else:
            print("Subject Fail")
            failed = True

percentage = total_marks / len(subjects)
cgpa = percentage / 9.5

print(f"\nTotal Marks: {total_marks}")
print(f"Percentage: {percentage}%")
print(f"CGPA: {round(cgpa, 1)}")

if failed:
    print("Overall Result: Failed")
else:
    print("Overall Result: Passed")
```
