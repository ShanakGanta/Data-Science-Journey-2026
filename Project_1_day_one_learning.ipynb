```python
                                          #Retail POS(Point Of Sale)
(product)=input("Enter the product: ")
(unit_price)=input("Enter the price: ")
(quantity)=input("Enter the Quantity: ")
unit_price=float(unit_price)
quantity=int(quantity)
sub_total=unit_price*quantity
tax=(sub_total*0.18)
discount_trigger=(tax>=100)
# Logic: (tax >= 100) becomes 1 if True, 0 if False.
# We multiply that 1 or 0 by the discount amount (e.g., 50).
actual_discount=discount_trigger*58
result=str(sub_total+tax-actual_discount)
print("Total Amount: ",result)
```
